```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dirs = ['simone-ms', 'giulia-ms']
dfs = []
```

```python
for dir in dirs:
    df = pd.read_csv(f'{dir}/results.txt', header=None)
    df['name'] = dir
    dfs.append(df)
```

```python
combined_df = pd.concat(dfs)
# combined_df.reset_index(drop=True, inplace=True)
```

```python
print(combined_df.info())
print(combined_df.describe().transpose())
```

```python
df = combined_df.groupby('name')
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Group the combined DataFrame by the 'file' column
grouped_df = combined_df.groupby('name')

# Plot histograms for each group
for file_name, group in grouped_df:
    plt.hist(group[0], bins=20, alpha=0.5, label=file_name)  # Assuming the values are in the first column (index 0)
    
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Values Grouped by File')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


# Create a box plot for each group
plt.figure(figsize=(20, 12))  # Adjust the figure size if needed
df.boxplot(column=0, by='name', patch_artist=True)

# Set plot labels and title
plt.xlabel('File')
plt.ylabel('Value')
plt.title('Box Plot of Values Grouped by File')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
from scipy.stats import wilcoxon
from tabulate import tabulate

# Group the combined DataFrame by the 'file' column
grouped_df = combined_df.groupby('name')

# Create a list to store values for each file
file_values = [grouped_df.get_group(file)[0].values for file in grouped_df.groups]

# Perform Wilcoxon signed-rank test between pairs of files
p_values = []
for i in range(len(file_values)):
    for j in range(i + 1, len(file_values)):
        _, p_value = wilcoxon(file_values[i], file_values[j])
        p_values.append((list(grouped_df.groups.keys())[i], list(grouped_df.groups.keys())[j], p_value))

# Create a table of p-values
table_headers = ['File 1', 'File 2', 'P-value']
table_data = [(file1, file2, p_value) for file1, file2, p_value in p_values]
table = tabulate(table_data, headers=table_headers, floatfmt=".2e", tablefmt="fancy_grid")

# Display the table
print(table)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon

# Group the combined DataFrame by the 'file' column
grouped_df = combined_df.groupby('name')

# Create a list to store values for each file
file_values = [grouped_df.get_group(file)[0].values for file in grouped_df.groups]
file_names = list(grouped_df.groups.keys())

# Perform Wilcoxon signed-rank test between pairs of files
num_files = len(file_values)
p_values_matrix = np.zeros((num_files, num_files))

for i in range(num_files):
    for j in range(i + 1, num_files):
        _, p_value = wilcoxon(file_values[i], file_values[j])
        p_values_matrix[i, j] = p_value
        p_values_matrix[j, i] = p_value

# Create a DataFrame with file names as indices
p_values_df = pd.DataFrame(p_values_matrix, index=file_names, columns=file_names)

# Print the matrix-style representation with file names
print("Comparison Matrix:")
print(p_values_df)
```

```python
# Assuming p_values_df is your DataFrame containing the comparison matrix

# Flatten the upper triangle of the comparison matrix (excluding the diagonal)
p_values_flat = p_values_df.where(np.triu(np.ones(p_values_df.shape), k=1).astype(bool)).stack()

# Sort the p-values in ascending order
sorted_p_values = p_values_flat.sort_values()

# Extract the pair(s) with the lowest p-value(s)
top_pairs = sorted_p_values.head(2)  # Change the number to extract more pairs if needed

# Display the top pairs
print("Pairs with the lowest p-values:")
print(top_pairs)
```
